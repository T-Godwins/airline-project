### Data Cleaning

```python
def clean_data(df):
    df.dropna(inplace=True)
    for col in df.select_dtypes(include=['object']).columns:
        df[col] = df[col].fillna('')
        df[col] = df[col].str.strip().str.lower()
        df['date'] = pd.to_datetime(df['date'])

    print("----Data Cleaned----\n")
    print(df.info())
    print(df.isnull().sum())
    return df

clean_data(delta_df)
```
